```python
import random
import matplotlib.pyplot as plt
import pandas as pd

def has_consecutive_kings(deck):
    for i in range(len(deck) - 1):
        if deck[i] == 'K' and deck[i + 1] == 'K':
            return True
    return False

def monte_carlo_simulation(num_simulations):
    deck = ['K'] * 4 + ['X'] * 48
    count = 0
    for _ in range(num_simulations):
        random.shuffle(deck)
        if has_consecutive_kings(deck):
            count += 1
    return count / num_simulations

if __name__ == "__main__":
    num_simulations_list = [10, 100, 1000, 10000, 100000, 1000000]
    results = []

    for num in num_simulations_list:
        prob = monte_carlo_simulation(num)
        results.append(prob)

    df = pd.DataFrame({"Number of Simulations": num_simulations_list, "Estimated Probability": results})
    print(df.to_string(index=False))

    plt.figure(figsize=(8, 5))
    plt.plot(num_simulations_list, results, marker='o', linestyle='-', color='b')
    plt.xscale("log")
    plt.xlabel("Number of Simulations")
    plt.ylabel("Estimated Probability")
    plt.title("Monte Carlo Simulation of Consecutive Kings")
    plt.grid(True)
    plt.show()
```
